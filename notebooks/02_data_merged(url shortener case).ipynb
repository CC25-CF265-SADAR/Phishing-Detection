```python
import pandas as pd
import numpy as np
import os
```

```python
path_to_folder = '../data/processed/'
df_pertama = pd.read_csv(os.path.join(path_to_folder, 'dataset_gabungan.csv'))
df_pertama
```

```python
path_to_folder = '../data/processed/'

df_kedua = pd.read_csv(os.path.join(path_to_folder, 'dataset_url_shortener_only.csv'))
df_kedua
```

```python
# 1. Dapatkan daftar URL yang akan dihapus dari data kedua
url_untuk_dihapus = df_kedua['url_original']

# 2. Filter df_pertama untuk menyimpan baris yang 'url_original'-nya
#    TIDAK ADA (~) di dalam daftar url_untuk_dihapus.
df_pertama_bersih = df_pertama[~df_pertama['url_original'].isin(url_untuk_dihapus)]

print("\n--- Data Pertama Setelah Baris yang Sama Dihapus ---")
df_pertama_bersih
```

```python
# Gabungkan (append) df_kedua ke df_pertama_bersih
df_final = pd.concat([df_pertama_bersih, df_kedua], ignore_index=True)

print("\n--- Hasil Akhir Setelah Digabung ---")
df_final
```

```python
# save the final DataFrame to a new CSV file
path_to_folder = '../data/processed/'

df_final.to_csv(os.path.join(path_to_folder, 'dataset_gabungan with_shortener_case.csv'), index=False)
```
