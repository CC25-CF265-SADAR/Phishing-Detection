```python
import pandas as pd

df_shortener = pd.read_csv("../../data/data playground/dataset_url_shortener.csv")
df_shortener
```

```python
kolom = df_shortener.iloc[:, 3:24].copy()
kolom
```

```python
nama_kolom = kolom.columns
nama_kolom
```

```python
# ambil kolom 1 dan 2

# ambil sampel 10 data aja
df_shortener_prep_sampel = df_shortener.copy()

df_shortener_prep=df_shortener_prep_sampel.iloc[:, 0:3]
df_shortener_prep
```

```python
# 1. Impor fungsi dari file .py Anda
# Pastikan file 'feature_extractor_shortener.py' ada di direktori yang sama
# atau di dalam PYTHONPATH.
from feature_extractor_shortener import extract_features
from tqdm import tqdm
import pandas as pd
# Hitung jumlah fitur yang diharapkan
jumlah_fitur = len(nama_kolom)

# 4. Iterasi, Ekstraksi, dan Pengumpulan Hasil
hasil_ekstraksi = []
# Menggunakan tqdm untuk melihat progress bar
for url in tqdm(df_shortener_prep['url_original'], desc="Mengekstrak Fitur URL"):
    try:
        # Panggil fungsi untuk mengekstrak fitur dari setiap URL
        fitur = extract_features(url)
        hasil_ekstraksi.append(fitur)
    except Exception as e:
        # Jika terjadi error tak terduga pada satu URL, catat error tersebut
        # dan isi dengan nilai default agar proses tidak berhenti dan data tidak bergeser.
        print(f"Error pada URL '{url}': {e}")
        hasil_ekstraksi.append([-1] * jumlah_fitur)

# 5. Buat DataFrame baru dari hasil ekstraksi
df_fitur = pd.DataFrame(hasil_ekstraksi, columns=nama_kolom)


# 6. Gabungkan DataFrame asli dengan DataFrame fitur
df_final = pd.concat([df_shortener_prep.reset_index(drop=True), df_fitur], axis=1)

```

```python
# 7. Simpan DataFrame akhir ke file CSV
df_final.to_csv("../../data/processed/dataset_url_shortener_only.csv", index=False)

df_final
```
